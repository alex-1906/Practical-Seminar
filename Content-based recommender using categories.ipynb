```python
import pandas as pd
```

```python
ratings_df = pd.read_csv('ratings.csv')
movies_df = pd.read_csv('movies.csv')
```

# Preprocessing

```python
movies_df['genres'] = movies_df.genres.str.split('|')
#movies_df.movieId = movies_df.movieId.astype('int32')
movies_df
```

```python

# First let's make a copy of the movies_df
movies_with_genres = movies_df.copy(deep=True)

# Let's iterate through movies_df, then append the movie genres as columns of 1s or 0s.
# 1 if that column contains movies in the genre at the present index and 0 if not.

x = []
for index, row in movies_df.iterrows():
    x.append(index)
    for genre in row['genres']:
        movies_with_genres.at[index, genre] = 1

# Confirm that every row has been iterated and acted upon
print(len(x) == len(movies_df))
movies_with_genres = movies_with_genres.fillna(0)
movies_with_genres.head(3)
```

```python
ratings_df.drop('timestamp',axis=1,inplace=True)
```

```python
ratings_df.head()
```

```python
movie_ratings = pd.merge(ratings_df,movies_df,on='movieId')
movie_ratings_user = movie_ratings.pivot_table(index='userId',columns='title',values='rating')
movie_ratings_user
```

```python
#movies of user 19
user = 19
user_profile = movie_ratings_user.iloc[user]
user_profile.dropna(inplace = True)
user_profile =pd.DataFrame(user_profile)
```

```python
#reverse mapping from movie name to movieId
indices = pd.Series(movies_df.index,index=movies_df['title'])
indices
```

```python
movies_with_genres.drop(columns=['movieId','title','genres'],inplace=True)
movies_with_genres.iloc[indices['Toy Story (1995)']]
```

```python


def create_user_profile(userId):
    user_profile = movie_ratings_user.iloc[userId]
    user_profile.dropna(inplace = True)
    user_ratings = user_profile
    liste = []
    for i in user_profile.index:
        liste.append(movies_with_genres.iloc[indices[i]])
    user_profile_df = pd.concat(liste,axis=1).T
    user_profile_df = user_profile_df.reset_index()
    user_profile_df.drop(columns=['movieId','genres','index'],inplace=True)
    return user_ratings,user_profile_df
ratings,profile = create_user_profile(23)
print('shape of ratings: ',ratings.T.shape)
print('shape of profile: ',profile.shape)
ratings
ratings_df = pd.DataFrame(list(zip(ratings.index,ratings.values)),columns=['title','rating'])
ratings_df
```

```python
userprof = profile.T.dot(ratings_df.rating)
userprof
```

```python
user_profile_df = pd.DataFrame(list(zip(userprof.index,userprof.values)),columns=['title','score'])
user_profile_df.drop(axis='index',index=0,inplace=True)
user_profile_df
```

```python
movies_with_genres.drop(columns=['movieId','title','genres'],inplace=True)
movies_with_genres
```

```python
print(movies_with_genres.shape)
print(user_profile_df.score.shape)
movies_with_genres.head(5)
```

```python
user_profile_df.head(5)
user_profile_df = user_profile_df.reset_index()
#user_profile.drop(columns=['index'],inplace=True)
user_profile_df.columns
```

```python
user_profile_df
```

```python
recommendations_table = movies_with_genres.dot(user_profile_df.score.to_numpy()) /(user_profile_df.score.sum())
recommendations_table
```

```python
recommendations_table.sort_values(ascending=False,inplace=True)
top_10_index = recommendations_table.index[:10].tolist()
print(top_10_index)
```

```python
top_10_movies = movies_df.iloc[top_10_index]
top_10_movies
```
